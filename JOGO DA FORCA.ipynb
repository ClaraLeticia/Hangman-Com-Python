## JOGO DA FORCA



#### Para desenvolver o jogo da forca em Python, você pode seguir os seguintes passos:


1- Definir a lista de palavras possíveis

2- Escolher uma palavra aleatória da lista

3- Criar uma lista vazia para armazenar as letras adivinhadas

4- Definir o número máximo de tentativas permitidas

5- Enquanto o número de tentativas não atingir o limite máximo:

a. Mostrar a palavra como uma série de underscores, com as letras adivinhadas preenchidas nos espaços corretos

b. Pedir ao jogador que adivinhe uma letra

c. Verificar se a letra adivinhada está na palavra

d. Se a letra adivinhada está na palavra, adicionar a letra à lista de letras adivinhadas e atualizar a exibição da palavra

e. Se a letra adivinhada não está na palavra, reduzir o número de tentativas restantes e exibir a mensagem "Letra incorreta. Tentativas restantes: [número de tentativas restantes]"

f. Verificar se todas as letras da palavra foram adivinhadas

g. Se todas as letras foram adivinhadas, exibir a mensagem "Você venceu!"

h. Se o número de tentativas restantes chegar a zero, exibir a mensagem "Você perdeu. A palavra era [palavra escolhida]" e encerrar o jogo. 

```python

##Definindo variaveis
palavra = "" ## Crio a variavel que vai armazenar a palavra da rodada
letras = [] ## Criando uma lista para armazenar as tentativas de letras
lista = [] ## Criando uma lista para armazenar as letras da palavra
maxten = 6 ## Número maximo vai ser 6


## Armazenando os valores do arquivo na lista
arquivo = open('Palavras.txt', 'r')
dados = arquivo.read()
listapalavras = dados.split("\n")


def bemvindo (): ## Função bem vindo de introdução
    escolherpalavra(listapalavras)
    global palavra
    print("Bem vindo!\n")
    print(f"Advinhe a palavra:\n")
    for i in range(0, len(palavra)):
        lista.append("_") ## Armazenando a quantidade de letras
    for i in range(0, len(palavra)):
        print(lista[i], end = " ") ## Parâmetro end é para escrever na mesma linha
    jogar() ## chamando a função jogar
    
def escolherpalavra(listapalavra):
    global palavra
    import random
    var = random.randint(0, len(listapalavras))
    palavra = listapalavras[var]
    
def jogar(): ## Função jogar
    global maxten
    global letras
    global lista
    while maxten != 0: ## Enquanto houver tentativas
        print(f"\n\n\nVocê tem {maxten} tentativas\n") 
        letra = input(f"Escolha uma letra: ")
        letras.append(letra) ## Adicionando a letra escolhida na lista
        verificar(letra) ## Chamando a função verificar
        mostrarletra() ## Chamando a função de mostrar as letras escolhidas
        ganhou(palavra) ## Verficando se ganhou
        if ganhou(palavra): 
            maxten = 0 ## Caso tenha ganhado definir maxten é igual a zero, e encerra as tentativas
            print("\n\n\nParabéns você ganhou!!!")
    
    if maxten == 0:
        print("\n\n\nA palavra anterior era: ", palavra)
        nov = input("\n\n\nDeseja jogar novamente?\n\n\n1- SIM\n2- NÃO\n")
        if nov == "1":
            maxten = 6
            letras = []
            lista = []
            bemvindo()
        else:
            print("Obrigada por jogar!!!")
       
        
        
def verificar(letra): ## Verificar se tem a letra na palavra
    global lista ## indicando que a lista que vamos trabalhar é a global
    global maxten
    if letra in palavra:
        for i in range(0, len(palavra)):
            if letra == palavra[i]: ## Verificando se existe
                lista[i] = f"{letra}" ## Caso exista a letra vai "tomar" a posição na lista
    else:
        maxten -= 1
        
        
        
           
    for i in range(0, len(palavra)):
        print(lista[i], end = " ")


    
def mostrarletra(): ## Função para mostrar as letras já escolhidas
    print("\n\n\nLetras escolhidas: ", end = " ") 
    for i in range(0, len(letras)):
        print(letras[i], end = " ")


def ganhou(palavra):
    global lista
    juncao = ''.join(lista)
    if juncao == palavra:
        return True

##MAIN
        
bemvindo()
```
